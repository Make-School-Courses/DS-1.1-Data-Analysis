## Question 2

What is the average amount of sugars based on manufacturer for cold cereals? (Indicated by "C" in the type column)? Support your answer with a graph or visualization. Hint: groupby, agg, and np.mean will be helpful here

```python
import pandas as pd
import seaborn as sns
import numpy as np

df = pd.read_csv("Datasets/cereal.csv")
df[df["type"] == "C"].groupby("mfr")["sugars"].agg(np.mean).plot(kind="bar")
```
